```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)
print(digits.target.shape)
```

## Input free Noise data

```python
def plot_digits(data):
    fig, axes = plt.subplots(8,8, figsize=(5,5),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest')
        #ax.text
plot_digits(digits.data)
```

## Create noise for your data

```python
rng = np.random.RandomState(101)
noise = rng.normal(digits.data,4)#you can vary the noise.
plot_digits(noise)
```

**The above images are noisy.**

```python
from sklearn.decomposition import PCA
pca = PCA(.4).fit(noise)
pca.n_components_
```

## reconstruction of the image from the dimensions.

```python
components = pca.transform(noise)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```
